```python
from urllib.request import urlopen, urlretrieve, Request
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#  Obtendo o HTML
url = "https://fbref.com/pt/equipes/03ff5eeb/2023/partidas/all_comps/gca/Cruzeiro-Historicos-dos-Jogos-Todos-os-campeonatos"
headers ={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36"}
 
req = Request(url,headers=headers)
response = urlopen(req)
html = response.read()
soup = BeautifulSoup(html,'html.parser')
 
#  Obtendo as TAGs de interesse da tabela
lista = soup.find('table')
```

```python
#  Ações e suas respectivas quantidades
registro = soup.findAll('th',class_='left') #  os registros da tabela
qtd_1 = range(int(len(registro)))
resumo = []
qtd_1
```

```python
#  Informações da primeira linha que não entram no loop
data =  lista.find('th',class_='left').getText()
horario = lista.find('td').contents[0]
```

```python
#  Para cada linha (i) em registro, obtenha os seguintes conteúdos:
for i in registro:
    cruzeiro ={}
    camp = horario.findNext('td').contents[0]
    rodada = camp.findNext('td').contents[0]
    dia = rodada.findNext('td').contents[0]
    local = dia.findNext('td').contents[0]
    resultado = local.findNext('td').contents[0]
    gp = resultado.findNext('td').contents[0]
    gc = gp.findNext('td').contents[0]
    oponente = gc.findNext('td').contents[0]
    sca = oponente.findNext('td').contents[0]
    passlive_sca = sca.findNext('td').contents[0]
    passdead_sca = passlive_sca.findNext('td').contents[0]
    to_sca = passdead_sca.findNext('td').contents[0]
    tc_sca = to_sca.findNext('td').contents[0]
    fitsp_sca = tc_sca.findNext('td').contents[0]
    def_action_sca = fitsp_sca.findNext('td').contents[0]
    gca = def_action_sca.findPrevious('td').findNext('td').contents[0]
    passlive_gca = gca.findNext('td').contents[0]
    passdead_gca = passlive_gca.findNext('td').contents[0]
    to_gca = passdead_gca.findNext('td').contents[0]
    tc_gca = to_gca.findNext('td').contents[0]
    fitsp_gca = tc_gca.findNext('td').contents[0]
    def_action_gca = fitsp_gca.findNext('td').contents[0]
    relatorio = def_action_gca.findNext('td').contents[0]
    
    #  Armazene as informações em cruzeiro da seguinte forma:
    cruzeiro['Data'] = data
    cruzeiro['Horário'] = horario
    cruzeiro['Campeonato'] = camp
    cruzeiro['Rodada'] = rodada
    cruzeiro['Dia'] = dia
    cruzeiro['Local']= local
    cruzeiro['Resultado'] = resultado
    cruzeiro['GP'] = gp
    cruzeiro['GC'] = gc
    cruzeiro['Oponente'] = oponente
    cruzeiro['SCA'] = sca
    cruzeiro['PassLive_sca'] = passlive_sca
    cruzeiro['PassDead_sca'] = passdead_sca
    cruzeiro['TO_sca'] = to_sca
    cruzeiro['TC_sca'] = tc_sca
    cruzeiro['Faltas_sca'] = fitsp_sca
    cruzeiro['Ações_Defensivas_sca'] = def_action_sca
    cruzeiro['GCA'] = gca
    cruzeiro['PassLive_gca'] = passlive_gca
    cruzeiro['PassDead_gca'] = passdead_gca
    cruzeiro['TO_gca'] = to_gca
    cruzeiro['TC_gca'] = tc_gca
    cruzeiro['Faltas_gca'] = fitsp_gca
    cruzeiro['Ações_Defensivas_gca'] = def_action_gca
    
    #  Armazene cruzeiro na lista vazia criada anteriormente
    resumo.append(cruzeiro)
    
    #  A última linha deve dar erro, pois não tem próxima etapa. Por isso, é necessária uma exceção:
    try:
        data = i
        horario = data.findNext('td').contents[0]
    except:
        print('Processo terminado.')
```

```python
#  Cria data frame com dados extraídos anteriormente
df = pd.DataFrame(resumo)
 
display(df)
```
